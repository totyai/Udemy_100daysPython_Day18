## Ways to import modules
- import turtle
    - This imports the whole package. And when we are using a method or attribute from it, we always need to call the package in
    - turtle.Turtle()
    - Reccomended to use if oen time you use a mothod
- from turtle import Turtle
    - This is making easier to access the attributes of the packages by importing only those that we need. And we can reference only those, no need to call the package again
    - Turtle()
- from turtle import *
    - This imports every modul, attribute seprately from the package, so no need to call in the package itself
    - It create a bit of confusion, on which package does a mehthod come from, so not usual
- import turtle as t - alias a modul
    - give the turtle modul a custom name, we define, so in the code it will be referenced by that name

- Sometime you need to **install** moduls, not jsut call them and import them.
    - These are libraries/packages that are not included by default in the python package library
    - These installs are found in the PyPi.
    - These installs are usually only installed *local environemnt*, basically with the projct files itself, but not across the device
    - This is because pyhton is working with Virtual Envirnments
        - To be able to use different versions of python across projects

## Challange 3
- Draw different shapes on each other, each with a random color
- Shapes: Triange, Square, Pentagon, Hexagon, Heptagon, Octogon, Nonagon, Decagon
- Each each of each shape is 100

```python
shapes = Turtle()

for i in range(3,11):
    colour = choice(COLORS)
    angle = 360/i
    shapes.pencolor(colour)
    for u in range(0,i):
        shapes.fd(100)
        shapes.right(angle)

screen = Screen()
screen.exitonclick()
        
```

## Challange 4
Draw a random walk
Create a randomized movement for the turtle.
It will be random which direction it goes, will be random, what colour it draws, but it iwll always go 10 pieces.
Also increase speed of drawing and line thickness compared to before

```python
random_walk = Turtle()
random_walk.pensize(15)
random_walk.speed(7)

directions = [random_walk.setheading(0), random_walk.setheading(90), random_walk.setheading(180), random_walk.setheading(270)]
for _ in range(100):
    random_walk.pencolor(choice(COLORS))
    random_walk.fd(10)
    choice(directions)

screen = Screen()
screen.exitonclick()
```
